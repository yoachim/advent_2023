```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('input', header=None)
```

```python
data
```

```python
lol = []
for line in data[0]:
    lol.append([char for char in line])
```

```python
np.unique(lol)
```

```python
# find the locations of "characters"

char_coords = []

chars = '*'

for i in range(len(lol)):
    for j in range(len(lol[0])):
        if lol[i][j] in chars:
            char_coords.append((i,j))
```

```python
char_coords
```

```python
arr = np.array(lol)

gear_spots = np.where(arr == '*')
```

```python
mask = np.zeros(np.array(lol).shape)
```

```python
def return_numbers(char_coords, arr_in):
    arr = arr_in.copy()
    mask = np.zeros(arr.shape)
    for cc in char_coords:
        for i in np.arange(-1,2):
            for j in np.arange(-1,2):
                mask[cc[0]+i, cc[1]+j] = 1

    grow_mask = mask*0

    digits = ['%i' % i for i in np.arange(10)]

    for dig in digits:
        grow_mask[np.where((mask > 0) & (arr == dig))] = 1

    previous_sum = -1
    while np.sum(grow_mask) != previous_sum:
        previous_sum = np.sum(grow_mask)
        for i,row in enumerate(grow_mask):
            hit = np.where(row > 0)[0]
            ahead = hit+1
            ahead = ahead[np.where(ahead < len(lol[0]))]
            for check in ahead:
                if lol[i][check] in digits:
                    grow_mask[i,check] = 1

            ahead = hit-1
            ahead = ahead[np.where(ahead > -1)]

            for check in ahead:
                if lol[i][check] in digits:
                    grow_mask[i,check] = 1


    arr[np.where(grow_mask == 0)] = ' '
    numbers = []
    for row in arr:
        row_str = ''
        for char in row:
            row_str += char
        blobs = row_str.split(' ')
        for blob in blobs:
            if blob != '':
                numbers.append(int(blob))
    return numbers
```

```python
return_numbers(char_coords, arr)
```

```python
return_numbers([char_coords[0]], arr)
```

```python
gear_ratios = []
for cc in char_coords:
    numbers = return_numbers([cc], arr)
    if len(numbers) == 2:
        gear_ratios.append(numbers[0]*numbers[1])
```
